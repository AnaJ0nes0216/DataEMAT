**Proposal: What time periods are more likely for an instagram user to view and interact with ads, recommended users, and posts in general. Will there be a connection between these data sets when the timestamps are put into consideration? I want to put these all into graphs and make a comparison. As a conclusion I want to be able to see from this data what specific times of the year I am more active on Instagram.**

```python
import pandas as pd
import json
from scipy import stats
from matplotlib import pyplot as plt
from datetime import datetime as dt
```

**I chose to use these data sets because they allow me to see when Instagram ads, posts, suggested accounts and videos were viewed. From those dates and times I can then determine what times of the year I was more actively viewing Instagram content. I downloaded my own Instagram data to my computer. Opening the Instagram data I downloaded from my computer and turning into json files. Then turn them into DataFrames so that I can see the timestamps for each set.**

```python
with open(r"C:\Users\Anamarie Jones\Downloads\ana_jones0216_20210907\ads_and_content\ads_viewed.json") as ad:
    ads = json.load(ad)
with open(r"C:\Users\Anamarie Jones\Downloads\ana_jones0216_20210907\ads_and_content\posts_viewed.json") as p:
    posts = json.load(p)
with open(r"C:\Users\Anamarie Jones\Downloads\ana_jones0216_20210907\ads_and_content\suggested_accounts_viewed.json") as s:
    accounts = json.load(s)
with open(r"C:\Users\Anamarie Jones\Downloads\ana_jones0216_20210907\ads_and_content\videos_watched.json") as v:
    videos = json.load(v)
```

***Within the DataFrames, I look through the keys to find where the timestamps are located. Each data set I found 'Time' under 'string_map_data' and singled out the timestamps. I then printed the length of data from the 'timestamp' column.***

```python
time = [x['string_map_data']['Time'] for x in ads['impressions_history_ads_seen']]
ads_viewed_time = pd.DataFrame(time)
print(len(ads_viewed_time))
```

```python
time2 = [x['string_map_data']['Time'] for x in posts['impressions_history_posts_seen']]
posts_viewed = pd.DataFrame(time2)
print(len(posts_viewed))
```

```python
time3 = [x['string_map_data']['Time'] for x in accounts['impressions_history_chaining_seen']]
accounts_viewed = pd.DataFrame(time3)
print(len(accounts_viewed))
```

```python
time4 = [x['string_map_data']['Time'] for x in videos['impressions_history_videos_watched']]
videos_viewed = pd.DataFrame(time4)
print(len(videos_viewed))
```

**Taking the DataFrames showing timestamps and turning them into readable dates and time. I use list comprehension and create a new column for the times in the DataFrames. This code was minimized to one line after having each seperate and now no longer prints the results full DataFrames.**

```python
ads_viewed_time['datetime'] = [dt.utcfromtimestamp(x) for x in ads_viewed_time['timestamp']]
posts_viewed['datetime'] = [dt.utcfromtimestamp(x) for x in posts_viewed['timestamp']]
accounts_viewed['datetime'] = [dt.utcfromtimestamp(x) for x in accounts_viewed['timestamp']]
videos_viewed['datetime'] = [dt.utcfromtimestamp(x) for x in videos_viewed['timestamp']]
```

**Using the newly added DataFrame column of dates I make seperate charts to compare the timestamps for each. I made these charts so that it can be easily seen what times throughout the year are most active. I did not include them on a single chart because the length of data documented for each is different and would not be visually appealing.**

```python
fig1, ax1 = plt.subplots(2, 2, figsize = (10, 10))
ax1[0, 0].plot(ads_viewed_time['datetime'], color = 'blue')
ax1[0, 0].set_title('Ads Viewed')
ax1[0, 1].plot(posts_viewed['datetime'], color = 'red')
ax1[0, 1].set_title('Posts Viewed')
ax1[1, 0].plot(accounts_viewed['datetime'], color = 'yellow')
ax1[1, 0].set_title('Suggested Accounts Viewed')
ax1[1, 1].plot(videos_viewed['datetime'], color = 'green')
ax1[1, 1].set_title('Videos Viewed')
plt.show()
```
